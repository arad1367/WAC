```python
# 1D array
array_1D = np.array([1, 3, 10])
array_1D

array_1D.ndim, array_1D.shape, array_1D.size, type(array_1D)

# 2D array
array_2D = np.array([
    [1, 3, 6],
    [5, 5, 5]
])

array_2D.ndim
array_2D

# 3D array
array_3D = np.array([[
    [1, 2, 3],
    [4, 4, 4]
]])

array_3D.ndim
```

#### Different ways to create an array

```python
from scipy.sparse import random
# Simple way
simple_array = np.array([1, 1, 1, 100])
simple_array

# Ones array
ones_array = np.ones((5, 5))
ones_array

# Zeros array
zero_array = np.zeros((3, 3))
zero_array

# Arange array
arange_array = np.arange(1, 100, 5)
arange_array

# linspace array
linspace_array = np.linspace(1, 10, 100)
linspace_array

# random array
random_array = np.random.randint(100, size=(5, 5))
random_array

# rand array
rand_array = np.random.rand(1000, 4)
rand_array
```
